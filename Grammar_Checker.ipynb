# Sentence to String(text)

```
sent_list=brown.sents()
a=len(brown.sents())
f=" "
for i in range(0,50):
  for j in sent_list[i]:
    f=f+" "+j.lower()
  f=f+"."
```

# Tokenization 

```
sentence=sent_tokenize(f)
word=word_tokenize(f)
```

# Stop-word Removal

```
stop_words = set(stopwords.words('english')) 
filtered_sentence1 = [] 
clean_sentences=[]

for i in range(len(sentence)):
  sent=sentence[i]
  word_tokens = word_tokenize(sent) 
  filtered_sentence1 = [w for w in word_tokens if not w in stop_words] 
  for w in word_tokens: 
    if w not in stop_words: 
        filtered_sentence1.append(w)
  clean_sentences.append(filtered_sentence1)

print(clean_sentences)
```

# Punctuation Removal

```
punctuation_tokenizer = RegexpTokenizer(r'\w+')

filtered_sentence2 = [] 
final_clean_sentences=[]
string=""

for i in range(len(clean_sentences)):
  sent=clean_sentences[i]
  cls= ' '.join(i for i in sent)
  #print(cls)
  result = punctuation_tokenizer.tokenize(cls)
  #print(result)
  final_clean_sentences.append(result)
  
print(final_clean_sentences)
#print(result)
```

```
final_clean_text1=[[' '.join(i)] for i in final_clean_sentences]
final_clean_text2=[' '.join(i) for i in final_clean_text1]
final_clean_text3=""
for i in final_clean_text2:
  final_clean_text3=final_clean_text3+i+". "
```

```
print(final_clean_text3)
```

# Dependacy Parsing using SpaCy and Capitalizing nouns

```
import spacy

nlp = spacy.load("en_core_web_sm")
nlp.max_length = 1600000
doc = nlp(final_clean_text3)
final_clean_text4=""
#doc[0]
for token in doc:
  #print(token.text)
  if(token.head.pos_=='NOUN'):
    final_clean_text4=final_clean_text4+token.text.capitalize()+" "
  else:
    final_clean_text4=final_clean_text4+token.text+" "
  #print(token.text, token.dep_, token.head.text, token.head.pos_, [child for child in token.children])
  #print(token.text, token.head.pos_)
```

```
print(final_clean_text4)
```

# POS Tag and Chunking 

```
def preprocess(sent):
    sent = nltk.word_tokenize(sent)
    sent = nltk.pos_tag(sent)
    return sent
  
sent = preprocess(final_clean_text4)
#sent
```

```
pattern = 'NP: {<DT>?<JJ>*<NN>}'
cp = nltk.RegexpParser(pattern)
cs = cp.parse(sent)
print(cs[0:100])
```
