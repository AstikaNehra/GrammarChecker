```
import pandas as pd
```

```
url_train = 'https://raw.githubusercontent.com/AstikaNehra/GrammarChecker/master/final_train.csv'
url_test = 'https://raw.githubusercontent.com/AstikaNehra/GrammarChecker/master/final_test.csv'
df_train = pd.read_csv(url_train,error_bad_lines=False, encoding='cp1252',header=None)
df_test = pd.read_csv(url_test,error_bad_lines=False, encoding='cp1252',header=None)
```

```
df_train
```

```
data_train=df_train.values
data_test=df_test.values
```

```
X_train = data_train[:,0]
Y_train = data_train[:,1]
X_test = data_test[:,0]
Y_test = data_test[:,1]
```

```
import nltk
from nltk.corpus import brown
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.tag import pos_tag  
nltk.download('brown')
nltk.download('punkt')
```

```
from nltk.corpus import stopwords  
nltk.download('stopwords')
from nltk.tokenize import RegexpTokenizer
nltk.download('averaged_perceptron_tagger')
```

```
sent_list_train=X_train.tolist()
f_train=""
for i in sent_list_train:
  #print(i)
  f_train=f_train+" "+i.lower()
  
sent_list_test=X_test.tolist()
f_test=""
for i in sent_list_test:
  #print(i)
  f_test=f_test+" "+i.lower()
```

```
sentence_train=sent_tokenize(f_train)
word_train=word_tokenize(f_train)

sentence_test=sent_tokenize(f_test)
word_test=word_tokenize(f_test)
```

```
stop_words = set(stopwords.words('english')) 
filtered_sentence1_train = [] 
clean_sentences_train=[]

for i in range(len(sentence_train)):
  sent=sentence_train[i]
  word_tokens = word_tokenize(sent) 
  filtered_sentence1_train = [w for w in word_tokens if not w in stop_words] 
  clean_sentences_train.append(filtered_sentence1_train)

filtered_sentence1_test = [] 
clean_sentences_test=[]

for i in range(len(sentence_test)):
  sent=sentence_test[i]
  word_tokens = word_tokenize(sent) 
  filtered_sentence1_test = [w for w in word_tokens if not w in stop_words] 
  clean_sentences_test.append(filtered_sentence1_test)
```

```
punctuation_tokenizer = RegexpTokenizer(r'\w+')

filtered_sentence2_train = [] 
final_clean_sentences_train=[]

for i in range(len(clean_sentences_train)):
  sent=clean_sentences_train[i]
  cls= ' '.join(i for i in sent)
  result = punctuation_tokenizer.tokenize(cls)
  final_clean_sentences_train.append(result)
  
print(final_clean_sentences_train)

filtered_sentence2_test = [] 
final_clean_sentences_test=[]

for i in range(len(clean_sentences_test)):
  sent=clean_sentences_test[i]
  cls= ' '.join(i for i in sent)
  result = punctuation_tokenizer.tokenize(cls)
  final_clean_sentences_test.append(result)
  
print(final_clean_sentences_test)
```

```
final_clean_text1_train=[[' '.join(i)] for i in final_clean_sentences_train]
final_clean_text2_train=[' '.join(i) for i in final_clean_text1_train]
final_clean_text3_train=""
for i in final_clean_text2_train:
  final_clean_text3_train=final_clean_text3_train+i+". "

final_clean_text1_test=[[' '.join(i)] for i in final_clean_sentences_test]
final_clean_text2_test=[' '.join(i) for i in final_clean_text1_test]
final_clean_text3_test=""
for i in final_clean_text2_test:
  final_clean_text3_test=final_clean_text3_test+i+". "
```
